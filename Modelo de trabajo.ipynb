# Construimos el modelo estadístico predictivo sobre violencia (OLS)
Intentaremos analizar cuáles son las mejores variables explicativas de la violencia contra la mujer ejercida por sus cónyuges. El modelo tentativo es el siguiente:

\begin{gather}
Violencia = β_{0}+β_{1}Reparto + β_{2}Edad + β_{3}Dificultad + β_{4}Estudio + β_{5}Parejas + β_{6}Redes + β_{7}Trabajo + β_{8}Sosten + β_{9}Violencia_infantil
\end{gather}

Donde Violencia es un indicador aditivo que mide la violencia multidimensional ejercida por el cónyuge sobre la mujer, Reparto es un indicador que mide qué tan repartidas están las labores domésticas en el hogar, Dificultad es una variable dummy que muestra si la mujer tiene alguna dificultad física o mental que la vuelva dependiente, Estudio es los años de estudio, Parejas es la cantidad de parejas previas tenidas, Redes es una dummy que muestra si la mujer participa o no en organizaciones locales, Trabajo es una variable que nos muestra si la mujer trabajó a cambio de una remuneración pecuniaria o en especie el último mes y Sosten es una variable dummy que muestra si la mujer participa o no en el sostén económico del hogar. 

Recordemos las variables:\
C1P208_A = edad\
C1P209 = dificultad física o mental\
C1P210_1 = Años de estudio\
C1P303 = número de parejas conyugales\
C1P310 = Trabajó el mes pasado

```r
data_violencia <- data[c("C1P208_A", "C1P209", "C1P210_1", "C1P303", "C1P310", "C1P318_2", "redes", 
"sosten", "control", "econ", "fisic", "sex", "violencia", "reparto", "C1P501", "C1P504", "C1P506")]
colnames(data_violencia) <- c("Edad", "Dificultad", "Estudio", "Parejas", "Trabaja", "Violencia infantil", 
"Redes", "Sosten", "Control", "Econ", "Fisic", "Sex", "Violencia", "Reparto", "Apoyo", "Institución", "Denunció")
```

```r
data_violencia2 <- data_violencia[complete.cases(data_violencia$Trabaja), ] # Hacemos un nuevo dataset excluyendo los valores perdidos
```

```r
modelo <- lm(data = data_violencia2, Violencia ~ Estudio + Reparto + Redes + Sosten + Edad + Dificultad + Trabaja + Parejas + `Violencia infantil`)
```

```r
summary(modelo)
```

### Test de likelihood ratio

*   $H_0$: Debe usar el modelo restringido
*   $H_a$: Debe usar el modelo irrestricto

En nuestro caso, la distinción entre ambos modelos será el uso de la variable violencia.

```r
install.packages("lmtest")
```

```r
library(lmtest)
```

```r
restringido <- glm(data = data_violencia2, Denunció ~ Estudio + Edad + Reparto + 
  Redes + Sosten + Sosten + Trabaja + Parejas + `Violencia infantil` + Violencia, 
  family = binomial(link = "logit"))
irrestricto <- glm(data = data_violencia2, Denunció ~ Estudio + Edad + Reparto + 
  Redes + Sosten + Sosten + Trabaja + Parejas + `Violencia infantil`, 
  family = binomial(link = "logit"))
```

```r
lrtest(restringido, irrestricto)
```

El p valor no es menor a 0.05, por lo que no hay evidencia estadística suficiente como para rechazar la hipótesis nula. Por tanto, nos quedamos con el modelo estricto.

### Test de Wald

```r
install.packages("aod")
```

```r
library(aod)
```

$H_0: Violencia = 0$

```r
wald.test(b = coef(restringido), Sigma = vcov(restringido), Terms = 10)
```

El p-valor es mayor a 0.05, por lo que no podemos rechazar la H0. No hay evidencia suficiente para afirmar que el coeficiente de esta variable es distinto de 0.
